```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

# Load the Iris dataset
iris = load_iris()

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
df

```

```python
fig, axes = plt.subplots(figsize=(20, 10), nrows=3, ncols=4)

# Define the colors for each column
colors = ['blue', 'orange', 'green', 'red']

# Iterate over the columns and plot the histograms, KDEs, and boxplots
col_no = 0
for col in df.columns[:-1]:
    sns.histplot(x=col, data=df, ax=axes[0, col_no], bins=20, color=colors[col_no])
    sns.kdeplot(x=col, data=df, ax=axes[1, col_no], color=colors[col_no])
    sns.boxplot(x=col, data=df, ax=axes[2, col_no], color=colors[col_no])
    col_no += 1

plt.tight_layout()
plt.show()
```

```python
labels = list(df["target"].value_counts().index)

plt.figure(figsize=(6, 8), dpi=80)
plt.pie(df["target"].value_counts(), autopct='%.2f%%', labels=labels)
plt.title("Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=80)
sns.heatmap(df.corr(), annot=True)
plt.show()
```
